### Part 1: MC Prediction

In this section, you will write your own implementation of MC prediction (for estimating the action-value function).  

We will begin by investigating a policy where the player _almost_ always sticks if the sum of her cards exceeds 18.  In particular, she selects action `STICK` with 80% probability if the sum is greater than 18; and, if the sum is 18 or below, she selects action `HIT` with 80% probability.  The function `generate_episode_from_limit_stochastic` samples an episode using this policy. 

The function accepts as **input**:
- `bj_env`: This is an instance of OpenAI Gym's Blackjack environment.

It returns as **output**:
- `episode`: This is a list of (state, action, reward) tuples (of tuples) and corresponds to $(S_0, A_0, R_1, \ldots, S_{T-1}, A_{T-1}, R_{T})$, where $T$ is the final time step.  In particular, `episode[i]` returns $(S_i, A_i, R_{i+1})$, and `episode[i][0]`, `episode[i][1]`, and `episode[i][2]` return $S_i$, $A_i$, and $R_{i+1}$, respectively.

````python
def generate_episode_from_limit_stochastic(bj_env):
    episode = []
    state = bj_env.reset()
    while True:
        probs = [0.8, 0.2] if state[0] > 18 else [0.2, 0.8]
        action = np.random.choice(np.arange(2), p=probs)
        next_state, reward, done, info = bj_env.step(action)
        episode.append((state, action, reward))
        state = next_state
        if done:
            break
    return episode
````

Now, you are ready to write your own implementation of MC prediction.  Feel free to implement either first-visit or every-visit MC prediction; in the case of the Blackjack environment, the techniques are equivalent.

Your algorithm has three arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `generate_episode`: This is a function that returns an episode of interaction.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

````python
env.action_space.n
````

````python
def mc_prediction_q(env, num_episodes, generate_episode, gamma=1.0):
    # initialize empty dictionaries of arrays
    returns_sum = defaultdict(lambda: np.zeros(env.action_space.n))
    N = defaultdict(lambda: np.zeros(env.action_space.n)) # keeps track of the number of appearances of a state-action pair
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        episode = generate_episode(env)
        
        running_discounted_reward = 0
        for time_step in episode[::-1]: # traverse the episodes in reverse order to obtain 
                                        # the cummulative discounted reward in each state
            state = time_step[0]
            action = time_step[1]
            reward = time_step[2]
            running_discounted_reward = gamma*running_discounted_reward+reward
            N[state][action]+=1 # implementing every visit monte-carlo
            returns_sum[state][action] += running_discounted_reward
            
    # having accumalated all the rewards from all the episodes we now compute the Q-values
    # In principle there might be state-action pairs that were never encountered during our interaction
    # In the present implementation we will only compute the Q-values of the state-action pairs
    # that were encountered at least once. The Q-value of the unencountered pairs will be left as zero
    # the encountered states are those that appear in the keys of the dictionary 'N'
    # the corresponding action was encountered if it's count is greater than 0
    
    for state in N.keys():
        for action in np.arange(2):
            if N[state][action]>0:
                Q[state][action] = returns_sum[state][action]/N[state][action]
    
    
    
    return Q
````

Use the cell below to obtain the action-value function estimate $Q$.  We have also plotted the corresponding state-value function.

To check the accuracy of your implementation, compare the plot below to the corresponding plot in the solutions notebook **Monte_Carlo_Solution.ipynb**.

````python
%%time
# obtain the action-value function
Q = mc_prediction_q(env, 500000, generate_episode_from_limit_stochastic)

# obtain the corresponding state-value function
V_to_plot = dict((k,(k[0]>18)*(np.dot([0.8, 0.2],v)) + (k[0]<=18)*(np.dot([0.2, 0.8],v))) \
         for k, v in Q.items())

# plot the state-value function
plot_blackjack_values(V_to_plot)
````

````python
# we can slightly improve our previous implementation of 'mc_prediction_q2'
# by constantly updating Q as we see an instance of state-action pair
# this is what was done in the solution
def mc_prediction_q2(env, num_episodes, generate_episode, gamma=1.0):
    # initialize empty dictionaries of arrays
    returns_sum = defaultdict(lambda: np.zeros(env.action_space.n))
    N = defaultdict(lambda: np.zeros(env.action_space.n)) # keeps track of the number of appearances of a state-action pair
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        episode = generate_episode(env)
        
        running_discounted_reward = 0
        for time_step in episode[::-1]: # traverse the episodes in reverse order to obtain 
                                        # the cummulative discounted reward in each state
            state = time_step[0]
            action = time_step[1]
            reward = time_step[2]
            running_discounted_reward = gamma*running_discounted_reward+reward
            N[state][action]+=1 # implementing every visit monte-carlo
            returns_sum[state][action] += running_discounted_reward
            Q[state][action] = returns_sum[state][action]/N[state][action]
 
    return Q
````

### Part 2: MC Control

In this section, you will write your own implementation of constant-$\alpha$ MC control.  

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.

The algorithm returns as output:

````python
# function to generate an episode by using the provided Q-table in an epsilon-greedy manner
def gen_episode(env, Qtable, epsilon):
    episode = []
    state = env.reset()
    while True:
        action_values = Qtable[state]
        
        # choose an action
        max_q = action_values.max() # the largest q-value for any action in the state
        max_action = np.ravel(np.argwhere(action_values == max_q)) # actions with the largest q-value
        remaining_actions = np.ravel(np.argwhere(action_values != max_q))
        
        if len(remaining_actions)!=0:
            # explore or exploit
            greedy = np.random.choice(2, p = [epsilon/2, 1-epsilon/2])
        
            if greedy == 1:
                # choose a greedy action
                action = np.random.choice(max_action)
            
            if greedy == 0:
                action = np.random.choice(remaining_actions)
        else:
            action = np.random.choice(max_action)
            
        # pass this action to the environment to recieve a reward and the next state
        next_state, reward, done, info = env.step(action)
        episode.append((state, action, reward))
        state = next_state
        if done:
            break
        
    return episode
````

````python
# function to compute the cummulative discounted rewards for the
# first visit of each state-action pair in an episode
def compute_discounted_rewards(episode, gamma):
    # to implement the 'first-visit' mc control solution, we will
    # keep a dictionary 'cummulative_rewards' which stores 
    # the cummulative_reward corresponding the firs-visit of each state-action pair
        
    cummulative_rewards = defaultdict(lambda : 0)
        
        
    # iterate through the time-steps of the episode in the reverse order
    # and compute the running_discounted_reward
    running_discounted_reward = 0
    for time_step in episode:
        state = time_step[0]
        action = time_step[1]
        reward = time_step[2]
        running_discounted_reward = gamma*running_discounted_reward + reward
        cummulative_rewards[(*state, action)] = running_discounted_reward
        
    return cummulative_rewards    
````

````python
def mc_control(env, num_episodes, alpha, gamma=1.0):
    nA = env.action_space.n
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        
        # generate an episode
        epsilon=max(1/i_episode, 0.1)
        episode = gen_episode(env, Q, epsilon)
        
        # compute the cummulative discounted reward for 
        # the first-visit of each state-action pair during the episode
        cummulative_rewards = compute_discounted_rewards(episode, gamma)
        
        # Use these cummulative_rewards to update the Q-table
        # recall the keys of cummulative_rewards are a tuple state,action 
        # the first 3 entries of the tuple give the state and the last entry gives the action
        for state_action, cum_reward in cummulative_rewards.items():
            state = state_action[:-1]
            action = state_action[-1]
            Q[state][action]+=alpha*(cum_reward - Q[state][action])
    
    # having iterated over the specified number of episodes, obtain 
    # the greedy policy from the Q-table
    policy = {}
    for state in Q:
        max_actions = np.ravel(np.argmax(Q[state])) # actions with maximum q-value
        # policy is given by choosing the action with maximum q-value
        policy[state] = np.random.choice(max_actions)
    
    return policy, Q
````

````python
# obtain the estimated optimal policy and action-value function
policy, Q = mc_control(env, 500000, 0.02)
````
